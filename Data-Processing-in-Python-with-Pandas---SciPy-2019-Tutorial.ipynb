__Objective 1:__ Write a function that takes a value of rate and parses out the total population. Note, the population is on the right hand side of the /.


```python
tbl3.dtypes
```

```python
# since rate is an "object" it is a string
```

```python
def extract_population(rate):
    population = rate.split('/')[1]
    return population
```

```python
# quick way to check to make sure the function won’t fail

assert extract_population("123/456") == "456"
```

```python
# to apply the function to the column

tbl3["rate"].apply(extract_population)
```

__Objective 2:__ Set the population to a new column.

```python
pop_column = tbl3["rate"].apply(extract_population)
```

# Part 5 - Encoding

```python
from sklearn import linear_model
```

```python
tips = sns.load_dataset("tips")
```

```python
tips.head()
```

```python
lr = linear_model.LinearRegression()
```

```python
lr.fit(X=tips[["total_bill","size"]], y=tips["tip"])
```

```python
lr.coef_

# LEARNING:

# For every one dollar increase in total_bill, given that 
# the size remains the same, the tip increases by 9 cents.
```

```python
lr.intercept_ 

# given that the total_bill is zero and no one shows up, the model is predicting a tip size of 67 cents
```

### Let's look at encoding our categorical variables now,

```python
pd.get_dummies(tips).head()
```

```python
tips_dummy = pd.get_dummies(tips, drop_first=True)
tips_dummy.head()
```

```python
lr = linear_model.LinearRegression()
lr.fit(X=tips_dummy.iloc[:, 2:], y=tips_dummy["tip"])
```

```python
lr.coef_

# Learning, if you are having dinner (versus not having lunch), then your tip is going
# to increase 0.48 cents assuming all the other features stay the same
```
